```python
# =================================================================================
# PIPELINES DE ENGENHARIA DE DADOS / MANIPULAÇÃO DE DATASETS COM PANDAS
# CONCEITO: Conversão de Queries do BigQuery diretamente para DataFrames e CSV
#
# CONTEXTO DE NEGÓCIO (Pedido do Gestor):
# "O time de finanças precisa realizar a conciliação bancária dos recursos críticos.
# Automatize a extração da query de custos acima da média, armazene os dados em
# uma estrutura de DataFrame do Pandas e exporte para um arquivo físico CSV,
# disponibilizando o arquivo pronto para download imediato do usuário."
# =================================================================================

from google.cloud import bigquery
import pandas as pd
import google.auth  # Biblioteca para capturar suas credenciais explicitamente
from google.colab import files
from google.colab import auth # Importando o módulo de autenticação do Colab

# Adicionando autenticação explícita para o usuário no Colab
print("Authenticating user...")
auth.authenticate_user()
print("User authenticated.")

# 1. CAPTURA DOS TOKENS: Puxa explicitamente a autenticação realizada na célula 1
credenciais, projeto_padrao = google.auth.default()

# 2. Inicializando o cliente injetando as suas credenciais de usuário logado
client = bigquery.Client(credentials=credenciais, project='gen-lang-client-0862909089')

# 3. Definindo a nossa query dinâmica de subquery
query = """
SELECT
  project.id AS projeto_id,
  service.description AS servico_descricao,
  cost AS custo_bruto
FROM
  `meu_finops.custos_nuvem_bruto_gcp`
WHERE
  cost > (SELECT AVG(cost) FROM `meu_finops.custos_nuvem_bruto_gcp`)
ORDER BY
  custo_bruto DESC;
"""

print("⏳ Buscando dados no BigQuery e convertendo para o Pandas...")

# 4. Executa a query e transforma o resultado direto em um DataFrame do Pandas
df_custos = client.query(query).to_dataframe()

# 5. Exibindo na tela o DataFrame formatado como uma tabela real de dados
print("\n📊 --- VISUALIZAÇÃO DA TABELA ESTRUTURADA (DATAFRAME) ---")
print(df_custos)
print("--------------------------------------------------------")

# 6. Salvando o DataFrame em um arquivo físico CSV
nome_arquivo = "relatorio_custos_criticos.csv"
df_custos.to_csv(nome_arquivo, index=False, sep=';', encoding='utf-8-sig')
print(f"✅ Arquivo '{nome_arquivo}' gerado com sucesso!")

# 7. Comando para forçar o navegador a fazer o download automático do arquivo
print("📥 Ready to download...")
files.download(nome_arquivo)
```
