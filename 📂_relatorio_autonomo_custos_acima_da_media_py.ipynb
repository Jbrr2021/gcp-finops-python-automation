# 🐍 Automação Corporativa com Python & BigQuery

## 🗂️ Conceito Técnico: Execução Programática de Queries via Google Client Library

### 💼 Contexto de Negócio (Pedido do Gestor):
> *"O nosso diretor de FinOps precisa analisar diariamente os recursos que operam acima da média de gastos, mas não possui acesso ao console do BigQuery. Crie um script em Python que automatize a conexão com o banco de dados, dispare a query de subquery analítica de forma transparente e imprima na tela um relatório limpo e formatado com os indicadores de Projeto, Serviço e Custo Real."*

---


```python
# =================================================================================
# ENGENHARIA DE DADOS AVANÇADA / AUTOMAÇÃO COM PYTHON & BIGQUERY
# CONCEITO: Conexão Nativa e Execução de Queries via Google Client Library
#
# CONTEXTO DE NEGÓCIO (Pedido do Gestor):
# "Automatize a execução da query de custos acima da média global utilizando Python,
# permitindo que o relatório seja extraído programaticamente sem necessidade de
# interação manual com o console do BigQuery."
# =================================================================================

# 1. Autenticando o seu usuário do Google dentro do Colab
from google.colab import auth
auth.authenticate_user()
print("✅ Autenticação realizada com sucesso!")

# 2. Importando a biblioteca oficial do BigQuery
from google.cloud import bigquery

# 3. Inicializando o cliente do BigQuery (substitua pelo ID do seu projeto se necessário)
# O Colab vai usar automaticamente o projeto padrão da sua conta Sandbox
client = bigquery.Client(project='gen-lang-client-0862909089')

# 4. Guardando a sua Super Query de Subquery dentro de uma variável de texto (String)
query = """
SELECT
  project.id AS projeto_id,
  service.description AS servico_descricao,
  cost AS custo_bruto
FROM
  `meu_finops.custos_nuvem_bruto_gcp`
WHERE
  cost > (SELECT AVG(cost) FROM `meu_finops.custos_nuvem_bruto_gcp`)
ORDER BY
  custo_bruto DESC;
"""

# 5. Mandando o Python executar a query dentro do BigQuery
print("⏳ Executando query no BigQuery...")
query_job = client.query(query)

# 6. Puxando os resultados e exibindo na tela com um loop (for)
resultados = query_job.result()

print("\n📊 --- RELATÓRIO: RECURSOS ACIMA DA MÉDIA GLOBAL ---")
for linha in resultados:
    print(# Usando f-string para formatar o texto bonitinho na tela
        f"Projeto: {linha.projeto_id} | "
        f"Serviço: {linha.servico_descricao} | "
        f"Custo: R$ {linha.custo_bruto:.2f}"
    )
print("----------------------------------------------------\n✅ Relatório finalizado!")
```
